```python
import cv2
import numpy as np

# خواندن تصویر
image = cv2.imread('D:/1.jpg')

# تبدیل تصویر به مقیاس خاکستری
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# اعمال فیلتر گوسی بر تصویر
blurred = cv2.GaussianBlur(gray, (5, 5), 0)

# تشخیص لبه‌ها با تکنیک Canny
edges = cv2.Canny(blurred, 50, 150)

# یافتن کانتورها
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# نمایش تصویر اصلی و کانتورها
cv2.imshow('Original Image', image)

# ساخت یک تصویر خالی برای نمایش کانتورها
contour_image = np.zeros_like(image)

# رسم کانتورها بر روی تصویر خالی
cv2.drawContours(contour_image, contours, -1, (0, 255, 0), 2)

# نمایش تصویر حاصل
cv2.imshow('Contours', contour_image)

# منتظر ماندن تا فشرده‌شدن یک کلید
cv2.waitKey(0)

# حذف پنجره‌ها
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# خواندن تصویر
image = cv2.imread('D:/1.jpg', cv2.IMREAD_GRAYSCALE)

# اعمال فیلتر گوسی بر تصویر
blurred = cv2.GaussianBlur(image, (5, 5), 0)

# تشخیص مرزها با استفاده از تکنیک Sobel
sobel_x = cv2.Sobel(blurred, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(blurred, cv2.CV_64F, 0, 1, ksize=3)

# محاسبه مقدار گرادیان مطلق
gradient_magnitude = np.sqrt(sobel_x**2 + sobel_y**2)

# تبدیل مقدار گرادیان به مقیاس uint8 (تصویر سیاه و سفید)
gradient_magnitude = np.uint8(255 * gradient_magnitude / np.max(gradient_magnitude))

# نمایش تصویر اصلی و تصویر تشخیص مرزها
cv2.imshow('Original Image', image)
cv2.imshow('Edge Detection (Sobel)', gradient_magnitude)

# منتظر ماندن تا فشرده‌شدن یک کلید
cv2.waitKey(0)

# حذف پنجره‌ها
cv2.destroyAllWindows()
```
